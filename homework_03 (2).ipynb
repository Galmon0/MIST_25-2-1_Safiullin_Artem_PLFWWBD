```python
import string
def count_word_frequency(text):
    table = str.maketrans('', '', string.punctuation) #string.punctuation — это готовая строка со всеми знаками препинания
    result = text.translate(table) #Возвращаем новую строку
    result = result.lower()
    result = result.split() #По словам
    res = dict()
    for i in result:
        if i not in res.keys():
            res[i] = 1
        else:
            res[i] += 1
    return res
text = "Это просто пример. Просто очень простой пример текста для примера."
frequency_dict = count_word_frequency(text)
print(frequency_dict)
```

```python
def calculate_price(price, discount=0, tax_rate=0.2):
    price1 = price*(100-discount)*0.01
    price2 = price1*(1+tax_rate)
    return price2
price = 100
discount = 10
new_price = calculate_price(price, discount)
print(new_price)
```

```python
from datetime import datetime
import re
def analyze_log(file_path):
    try:
        with open(file_path, 'r') as file:
            text = file.read()
    except FileNotFoundError:
        return {}
    entries = text.split('.')
    counts = {'INFO': 0, 'WARNING': 0, 'ERROR': 0}
    min_date = None
    max_date = None
    for entry in entries:
        entry = entry.strip() 
        if not entry:
            continue
        try:
            parts = entry.split(',')
            if len(parts) < 3:
                raise IndexError
            date_pattern = r'\d{4}-\d{2}-\d{2} \d{2}:\d{2}:\d{2}'
            match = re.search(date_pattern, parts[0])
            if not match:
                raise ValueError
            date_time_str = match.group()#Извлекаем дату
            level = parts[1].strip()
            message = parts[2].strip()
            date_obj = datetime.strptime(date_time_str, "%Y-%m-%d %H:%M:%S") 
            if level in ('INFO', 'WARNING', 'ERROR'):
                counts[level] += 1
                if min_date is None or date_obj < min_date:
                    min_date = date_obj
                if max_date is None or date_obj > max_date:
                    max_date = date_obj
        except (IndexError, ValueError):
            continue
    if min_date is None or max_date is None:
        return {'counts': counts, 'time_range': None}
    min_date_str = min_date.strftime("%Y-%m-%d %H:%M:%S")
    max_date_str = max_date.strftime("%Y-%m-%d %H:%M:%S")
    return {'counts': counts, 'time_range': (min_date_str, max_date_str)}

result = analyze_log('log.txt')
print(result)
```
