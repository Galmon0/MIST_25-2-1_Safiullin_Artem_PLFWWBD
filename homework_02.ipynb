```python
#Задача 1 "Классический FizzBuzz"
```

```python
N = int(input("Введите целое положительное число"))
if N >= 1:
    for i in range(1,N+1):
        if i % 3 == 0:
            if i % 5 == 0:
                print("FizzBuzz",end = " ")
            else:
                print("Fizz",end = " ")
        elif i % 5 == 0:
            print("Buzz",end = " ")
        else:
            print(i,end = " ")
else:
    print("Число не положительное")
```

```python
#Задача 2 "Интерактивный менеджер списка покупок"
```

```python
shopping_list = []
while True:
    print("1. Добавить")
    print("2. Показать")
    print("3. Выйти")
    a = int(input("Введите число"))
    if a == 1:
        b = input("Введите название продукта")
        shopping_list.append(b)
    elif a == 2:
        if len(shopping_list) == 0:
            print("Список покупок пуст.")
        else:
            for i in range(len(shopping_list)):
                print(i+1,". ",shopping_list[i], sep = "")
    elif a == 3:
        print("Программа завершена")
        break
    else:
        print("Выберите 1-3")
```
